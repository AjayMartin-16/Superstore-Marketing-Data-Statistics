## Task 3 - Probability Distributions üé≤

- **Identify variables that could follow specific probability distributions (e.g., Binomial, Normal). Calculate probabilities and expected values based on these distributions.**



**Deliverables**:

- **Determination of suitable probability distributions for relevant variables and corresponding calculated probabilities and expected values.**:

  

```python
from scipy.stats import binom

# Assuming `Response` has n=1 trials (single trial binary data)
p = df1['Response'].mean()  # Probability of success
probability = binom.pmf(k=1, n=1, p=p)
print("Probability of success (Response=1):", probability)
```

```python
from scipy.stats import norm

mean_income = df1['Income'].mean()
std_income = df1['Income'].std()

# Probability of income between 700000 and 90000
p = norm.cdf(700000, loc=mean_income, scale=std_income) - norm.cdf(90000, loc=mean_income, scale=std_income)
print("Probability of income between 700000 and 90,000:", p)
```

```python
from scipy.stats import poisson

lambda_visits = df1['NumWebVisitsMonth'].mean()
probability = poisson.pmf(k=8, mu=lambda_visits)  # Probability of 5 visits in a month
print("Probability of exactly 5 web visits in a month:", probability)
```

```python
from scipy.stats import shapiro

stat, p = shapiro(df1['Income'])
if p > 0.05:
    print("Income follows a normal distribution.")
else:
    print("Income does not follow a normal distribution.")
```

**Methods used for calculation:**

*   Finding binomial distribution for Response
*   Finding normal distribution for income
*   Finding poission distributioin for number of visits on the webpage
*   To check whether the income follows normal distribution using shapiro



## Task 4: Insights and Customer Segmentation üìà

- **Explore relationships between customer characteristics and spending habits. Segment customers based on their behaviors and characteristics.**

**Deliverables**:

- **Key insights regarding relationships between variables and distinct customer segments based on behaviors.**

  

```python
import seaborn as sns
import matplotlib.pyplot as plt

columns_to_analyze = ['Kidhome', 'Teenhome', 'Recency', 'MntWines', 'MntFruits',
                      'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts',
                      'MntGoldProds', 'NumDealsPurchases', 'NumWebPurchases',
                      'NumCatalogPurchases', 'NumStorePurchases',
                      'NumWebVisitsMonth', 'Response', 'Complain', 'Log_Income']


corr_matrix = df1[columns_to_analyze].corr()

plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()

```

```python
sns.scatterplot(data=df1, x='Income', y='MntWines')
plt.title('Income vs. Wine Spending')
plt.show()
```

```python
sns.boxplot(data=df1, x='Marital_Status', y='MntMeatProducts')
plt.title('Meat Spending by Marital Status')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

features = ['Income', 'MntWines', 'MntMeatProducts', 'NumWebPurchases', 'NumWebVisitsMonth']
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df1[features])
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4, random_state=42)
df1['Cluster'] = kmeans.fit_predict(df_scaled)
```

```python
cluster_summary = df1.groupby('Cluster').mean(numeric_only=True)
print(cluster_summary)
```

**High Spenders:**

 Cluster with high income and high spending on luxury products (MntWines, MntGoldProds).
**Budget-Conscious Shoppers:**

Cluster with low income, high NumDealsPurchases.

**Engaged Online Shoppers:**

Cluster with high NumWebPurchases and frequent NumWebVisitsMonth.
Actionable Insights:

Tailor marketing campaigns to target high-spending clusters.
Offer discounts to budget-conscious shoppers to drive loyalty.
